### **Gaussian Distribution**

```python
help(smart_dist.Gaussiandistribution)
```

```python
#Import Gaussian Class from Gaussiandistribution Module from the smart_dist Package
from smart_dist.Gaussiandistribution import Gaussian
```

```python
#Initialise a Gaussian Instance
Gauss = Gaussian()
Gauss
```

```python
#Creating a sample file for gaussian distribution

fw = open('gau_test.txt', 'w')

prob_array = np.asanyarray([i for i in range(2,14, 2)] + [r for r in range(10, 0, -2)])
prob = prob_array/sum(prob_array)

gau_flens = [fw.write(f'{v}\n') for v in np.random.choice(range(0,11), 20, p=prob)]
fw.close()
```

#### **Creating Gaussian Distribution from file Data**

```python
Gauss
```

```python
Gauss.read_data_file('gau_test.txt')
sorted(Gauss.data)
```

```python
Gauss.calculate_mean()
```

```python
Gauss.calculate_stdev()
```

```python
Gauss
```

#### **Finding probability of a number occuring**

```python
#For reference
sorted(Gauss.data)
```

```python
#Finding probablity of a number occuring 

nums = [i for i in range(1, 20, 1)]

num_prob = [ (num, round(Gauss.pdf(num) * 100, 2)) for num in nums ]


for num, prob in sorted(num_prob, key = lambda x: x[1], reverse = True):
    print(f'{num} -> {prob}%')
```

#### **Plotting distribution**

```python
Gauss.plot_histogram()
```

```python
Gauss.plot_histogram_pdf(n_spaces=len(prob_array))
```

#### **Adding Gaussian Distributions**

```python
gau1 = Gaussian(3, 5)
gau1
```

```python
gau2 = Gaussian(7, 10)
gau2
```

```python
gau3 = gau1 + gau2
gau3
```
